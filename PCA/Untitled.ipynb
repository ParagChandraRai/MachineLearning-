```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Given samples
samples = np.array([[4, 11], [8, 4], [13, 5], [7, 14]])

# Apply PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(samples)

# Display the results
print("Original Samples:")
print(samples)
print("\nPCA Result:")
print(pca_result)

# Plot the original and PCA-transformed samples
plt.figure(figsize=(8, 4))

# Original samples
plt.subplot(1, 2, 1)
plt.scatter(samples[:, 0], samples[:, 1], c='blue', label='Original Samples')
plt.title('Original Samples')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

# PCA-transformed samples
plt.subplot(1, 2, 2)
plt.scatter(pca_result[:, 0], pca_result[:, 1], c='red', label='PCA Result')
plt.title('PCA Result')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given samples
samples = np.array([[4, 11], [8, 4], [13, 5], [7, 14]])

# Step 1: Center the data
mean_vec = np.mean(samples, axis=0)
centered_samples = samples - mean_vec

# Step 2: Calculate the covariance matrix
cov_matrix = np.cov(centered_samples, rowvar=False)

# Step 3: Obtain eigenvectors and eigenvalues
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 4: Sort eigenvectors by eigenvalues (descending order)
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvectors = eigenvectors[:, sorted_indices]

# Step 5: Project the data onto the principal components
pca_result = np.dot(centered_samples, eigenvectors)

# Display the results
print("Original Samples:")
print(samples)
print("\nPCA Result:")
print(pca_result)

# Plot the original and PCA-transformed samples
plt.figure(figsize=(8, 4))

# Original samples
plt.subplot(1, 2, 1)
plt.scatter(samples[:, 0], samples[:, 1], c='blue', label='Original Samples')
plt.title('Original Samples')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

# PCA-transformed samples
plt.subplot(1, 2, 2)
plt.scatter(pca_result[:, 0], pca_result[:, 1], c='red', label='PCA Result')
plt.title('PCA Result')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()

plt.tight_layout()
plt.show()
```
