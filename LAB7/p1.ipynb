**Sample data (D1): Perform k-means clustering manually with k=2 on the given example and
plot the observations, label the cluser**s

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
X1 = np.array([1, 1, 0, 5, 6, 4])
X2 = np.array([4, 3, 4, 1, 2, 0])
data = np.column_stack((X1, X2))

# Initial centroids
c1 = np.array([0, 4])
c2 = np.array([4, 0])

# Number of iterations
max_iters = 10

# K-Means algorithm
for iteration in range(max_iters):
    # Assignment step
    distances_to_c1 = np.linalg.norm(data - c1, axis=1)
    distances_to_c2 = np.linalg.norm(data - c2, axis=1)
    cluster_assignment = np.where(distances_to_c1 < distances_to_c2, 0, 1)

    # Update step
    c1 = np.mean(data[cluster_assignment == 0], axis=0)
    c2 = np.mean(data[cluster_assignment == 1], axis=0)

# Final plot
plt.scatter(data[:, 0], data[:, 1], c=cluster_assignment, cmap='viridis', label='Data Points')
plt.scatter([c1[0], c2[0]], [c1[1], c2[1]], c='red', marker='X', label='Centroids')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('K-Means Clustering (k=2)')
plt.legend()
plt.show()
```
