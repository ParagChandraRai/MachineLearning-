<font color='red' size='5'>winequality Dataset</font>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suppress FutureWarnings
warnings.filterwarnings("ignore", category=FutureWarning)

df = pd.read_csv("winequality-red.csv",delimiter=";")
df
```


<font color='red' size='5'>1. Histograms</font>

```python
# Example: Histogram for 'fixed acidity'
plt.figure(figsize=(8, 4))
sns.histplot(df['fixed acidity'], bins=20, kde=True)
plt.title('Distribution of Fixed Acidity')
plt.xlabel('Fixed Acidity')
plt.ylabel('Frequency')
plt.show()
```


<font color='red' size='5'>2. Pair Plots</font>

```python
# Example: Pair plot for selected numerical features
selected_features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar']
sns.pairplot(df[selected_features])
plt.show()
```


<font color='red' size='5'>3.Box Plots</font>

```python
# Example: Box plot for 'fixed acidity' by 'quality'
plt.figure(figsize=(8, 4))
sns.boxplot(x='quality', y='fixed acidity', data=df)
plt.title('Box Plot of Fixed Acidity by Quality')
plt.xlabel('Quality')
plt.ylabel('Fixed Acidity')
plt.show()
```


 
<font color='red' size='5'>4. Scatter Plots</font>

```python
# Example: Scatter plot between 'alcohol' and 'pH'
plt.figure(figsize=(8, 4))
sns.scatterplot(x='alcohol', y='pH', data=df)
plt.title('Scatter Plot between Alcohol and pH')
plt.xlabel('Alcohol')
plt.ylabel('pH')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suppress FutureWarnings
warnings.filterwarnings("ignore", category=FutureWarning)

df = pd.read_csv("Housing.csv",)
df
```


<font color='red' size='5'>1. Histograms</font>

```python
# Example: Histogram for 'price'
plt.figure(figsize=(8, 4))
sns.histplot(df['price'], bins=20, kde=True)
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```


<font color='red' size='5'>2.Count plot</font>

```python
# Example: Count plot for 'furnishingstatus'
plt.figure(figsize=(8, 4))
sns.countplot(x='furnishingstatus', data=df)
plt.title('Count Plot of Furnishing Status')
plt.xlabel('Furnishing Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```



<font color='red' size='5'>3.Box Plots</font>

```python
# Example: Box plot for 'price' by 'furnishingstatus'
plt.figure(figsize=(8, 4))
sns.boxplot(x='furnishingstatus', y='price', data=df)
plt.title('Box Plot of Price by Furnishing Status')
plt.xlabel('Furnishing Status')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```


<font color='red' size='5'>4.Scatter Plots</font>

```python
# Example: Scatter plot between 'area' and 'price'
plt.figure(figsize=(8, 4))
sns.scatterplot(x='area', y='price', data=df)
plt.title('Scatter Plot between Area and Price')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```
