```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you have the DataFrame df_no_outlier
# Extract features (X) and target (Y)
X = df_no_outlier.drop('X5', axis='columns').values
Y = df_no_outlier['X5'].values

# Normalize features (optional but can help with convergence)
X_normalized = (X - X.mean(axis=0)) / X.std(axis=0)

# Add a bias term to the features
X_normalized = np.column_stack((np.ones(X_normalized.shape[0]), X_normalized))

# Initialize weights
theta = np.zeros(X_normalized.shape[1])

# Hyperparameters
learning_rate = 0.01
iterations = 1000

# Gradient Descent
for _ in range(iterations):
    # Predictions
    predictions = np.dot(X_normalized, theta)
    
    # Errors
    errors = predictions - Y
    
    # Update weights
    gradient = np.dot(X_normalized.T, errors) / len(Y)
    theta -= learning_rate * gradient

# Print the learned coefficients
print("Learned Coefficients:")
print(theta)

# Plot the actual vs predicted points
plt.scatter(Y, predictions, c='blue', label='Actual vs Predicted')
plt.plot([min(Y), max(Y)], [min(Y), max(Y)], linestyle='--', color='red', linewidth=2, label='Ideal Line')
plt.xlabel('Actual Average Points')
plt.ylabel('Predicted Average Points')
plt.title('Actual vs Predicted Average Points')
plt.legend()
plt.show()
```
